# Desaturation Filter

Saturation in the context of images refers to the intensity or purity of colors. It affects how vivid or dull the colors in an image appear.

1. Saturation:
- High Saturation: Colors become more vibrant and intense, making them appear richer and more vivid.
Low Saturation: Colors become muted or faded. As saturation decreases, the image becomes more "washed out."
- Zero Saturation: The image is completely grayscale, meaning all the colors have been removed, leaving only shades of gray.

2. Desaturation:
- Desaturation is the process of reducing the saturation of an image. When you apply desaturation, the colors become less vivid, and they move closer to grayscale.
- A fully desaturated image (saturation = 0) contains no color information, just varying levels of brightness in black and white.

```python
import cv2
```

```python
# Load the image
image = cv2.imread('./images/img.jpg')

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Convert grayscale image back to BGR (so we can blend with the original)
gray_image_bgr = cv2.cvtColor(gray_image, cv2.COLOR_GRAY2BGR)

# Set desaturation level (0.0 = grayscale, 1.0 = original image)
desaturation_level = 0.5  # Adjust this value (between 0.0 to 1.0)

# Blend the original and grayscale images
desaturated_image = cv2.addWeighted(image, desaturation_level, gray_image_bgr, 1 - desaturation_level, 0)

# Show the result
cv2.imshow('Desaturated Image', desaturated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
