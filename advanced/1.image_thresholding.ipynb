# Thresholding
Image thresholding is a technique used in image processing to create a binary image from a grayscale image. It converts pixels in the grayscale image to either black or white based on a threshold value. Pixels with values higher than the threshold are set to one value (typically white), and those with values lower are set to another (typically black). This process is useful for isolating objects from the background, which is important in various applications, such as object detection, segmentation, and computer vision.
Types of Thresholding

- Global Thresholding: A single threshold value is used for the entire image.
- Adaptive Thresholding: The threshold value is calculated for smaller regions of the image, which can be useful for images with varying lighting.
- Otsuâ€™s Thresholding: A method to automatically determine the optimal threshold value by minimizing intra-class variance.


### Implementation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the image
image = cv2.imread('./images/lena.png', cv2.IMREAD_GRAYSCALE)
image2 = cv2.imread('./images/face.jpg', cv2.IMREAD_GRAYSCALE)
```

**Global Thresholding**  
Here, 127 is the threshold value, and pixels above this value are set to 255 (white).

```python
_, global_thresh = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY) 
```

**Adaptive Thresholding**  
- cv2.ADAPTIVE_THRESH_GAUSSIAN_C: The threshold value is a weighted sum of the neighborhood values minus the constant C. This method usually gives better results, especially in images with varying lighting.
- Block Size = 11 (This is the size of the neighborhood area used to calculate the threshold value.)
- C = 2 (This is a constant that is subtracted from the mean or weighted sum calculated in the neighborhood (depending on the chosen adaptive method).)

```python
adaptive_thresh = cv2.adaptiveThreshold(image, 255,
                                         cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                         cv2.THRESH_BINARY,
                                         11, 2)
```

**Otsu's Thresholding**  
The cv2.threshold() function with cv2.THRESH_OTSU automatically finds the best threshold value.

```python
_, otsu_thresh = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

## Plotting the Results

```python
# Plotting the results
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(global_thresh, cmap='gray')
plt.title('Global Thresholding')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(adaptive_thresh, cmap='gray')
plt.title('Adaptive Thresholding')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(otsu_thresh, cmap='gray')
plt.title("Otsu's Thresholding")
plt.axis('off')

plt.tight_layout()
plt.show()
```
