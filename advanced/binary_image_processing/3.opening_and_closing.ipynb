# Opening and Closing

## 1. Opening
**Definition**: Opening is an erosion operation followed by a dilation.  
**Purpose**: It removes small objects or noise from the foreground (usually white) while preserving the shape and size of larger objects in the image.  
**Effect**: Smooths the contour of objects, breaks narrow connections, and eliminates thin protrusions.

Performing an opening operation allows us to remove small blobs from an image: first an erosion is applied to remove the small blobs, then a dilation is applied to regrow the size of the original object.

### IMPLEMENTATION

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Read the image in grayscale
image = cv2.imread('./images/ero.png', cv2.IMREAD_GRAYSCALE)
```

```python
#kernel
kernelSizes = [(3, 3), (5, 5), (7, 7)]
```

```python
plt.figure(figsize=(12, 12))  # Define the figure size
num_kernels = len(kernelSizes)  # Number of kernel sizes

for i, kernelSize in enumerate(kernelSizes):
    # Step 4: Create a rectangular kernel
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, kernelSize)
    
    # Step 5: Apply the "opening" operation
    opening = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)
    
    # Step 6: Plot the original image and the result for each kernel size
    plt.subplot(num_kernels, 2, 2*i+1)
    plt.imshow(image, cmap='gray')
    plt.title(f"Original Image")
    plt.axis('off')  # Hide axis for better visualization
    
    plt.subplot(num_kernels, 2, 2*i+2)
    plt.imshow(opening, cmap='gray')
    plt.title(f"Opening: ({kernelSize[0]}, {kernelSize[1]})")
    plt.axis('off')

# Display the plots
plt.tight_layout()
plt.show()
```

## 2. Closing
**Definition**: Closing is a dilation operation followed by an erosion.  
**Purpose**: It fills small holes and gaps within objects and connects nearby objects.  
**Effect**: Smooths the contours, fuses narrow breaks, and eliminates small dark spots on the objects.  

```python
plt.figure(figsize=(12, 12))  # Define the figure size
num_kernels = len(kernelSizes)  # Number of kernel sizes

for i, kernelSize in enumerate(kernelSizes):
    # Step 4: Create a rectangular kernel
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, kernelSize)
    
    # Step 5: Apply the "closing" operation
    closing = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)
    
    # Step 6: Plot the original image and the result for each kernel size
    plt.subplot(num_kernels, 2, 2*i+1)
    plt.imshow(image, cmap='gray')
    plt.title(f"Original Image")
    plt.axis('off')  # Hide axis for better visualization
    
    plt.subplot(num_kernels, 2, 2*i+2)
    plt.imshow(closing, cmap='gray')
    plt.title(f"Closing: ({kernelSize[0]}, {kernelSize[1]})")
    plt.axis('off')

# Display the plots
plt.tight_layout()
plt.show()
```
